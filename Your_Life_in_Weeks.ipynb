# **Solution**

```python
age = input()
# ðŸš¨ Don't change the code above ðŸ‘†
# Write your code below this line ðŸ‘‡
age = int(age)
years = 90 - age
no_of_weeks_in_year = 52
time_left = years * no_of_weeks_in_year
print(f"You have {time_left} weeks left.")
```
